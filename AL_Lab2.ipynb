MoveGen

```python
class Graph:
    def __init__(self):
        self.graph = {}
    def add_edge(self, node, neighbors):
        self.graph[node] = neighbors
def movegen(graph, node):
    if node in graph:
        return graph[node]
    else:
        return []
if __name__ == "__main__":
    graph = Graph()
    while True:
        node = input("Enter a node (or 'q' to quit): ")
        if node == 'q':
            break
        neighbors = input("Enter neighbors separated by commas: ").split(',')
        graph.add_edge(node.strip(), [neighbor.strip() for neighbor in neighbors])
    nodes = list(graph.graph.keys())
    for node in nodes:
        moves = movegen(graph.graph, node)
        print(f"Possible moves from {node} are: {moves}")
```

RemoveSeen

```python
def main():
    Open = []
    Closed = ["C", "D", "I", "J"]
    parent_node = input("Enter the parent node: ")
    while True:
        child_nodes = input("Enter child nodes separated by spaces ( 'q' to end): ")
        if child_nodes.lower() == 'q':
            break
        child_nodes = child_nodes.split()
        for child in child_nodes:
            if child not in Closed:
                Open.append(child)
                Closed.append(child)
            else:
                print("Discarding node:", child)
    while Open:
        current_node = Open.pop()
        print("Visiting node:", current_node)
if __name__ == "__main__":
    main()
```
