```python
import math
def haversine_distance(lat1, lon1, lat2, lon2):
  R = 6371.0
  lat1_rad = math.radians(lat1)
  lon1_rad = math.radians(lon1)
  lat2_rad = math.radians(lat2)
  lon2_rad = math.radians(lon2)

  dlon = lon2_rad - lon1_rad
  dlat = lat2_rad - lat1_rad

  a = math.sin(dlat / 2)**2 + math.cos(lat1_rad) * math.cos(lat2_rad) * math.sin(dlon / 2)**2
  C = 2 * math.atan2(math.sqrt(a), math.sqrt(1-a))

  distance = R * C
  return distance

locations = {
    'DJ Sanghvi college': (19.1048, 72.8375),
    'mithibai': (19.1075, 72.8377),
    'cooper hospital': (19.8598, 72.8387),
    'juhu circle': (19.1873, 72.8276),
    'irla petrol pump': (19.1868, 72.8381),
    'nanavati': (19.1026, 72.8385),
    'jw marriot-airport': (19.1123, 72.8614),
    'andheri station': (19.1197, 72.8464),
    'santacruz aeroplane park': (19.8994, 72.8539),
    'airport-andheri': (19.1191, 72.8482),
    'lokhandwaala market': (19.1308, 72.8297),
    'versova telephone exchange': (19.1317, 72.8141),
    'The Club-andheri': (19.1155, 72.8424),
    'BKC-JIO mall': (19.8671, 72.8679),
    'Worli Sealink': (19.0227, 72.8174),
    'Bandra-linking road': (19.0596, 72.8295),
    'Jaslok hospital': (18.9730, 72.8126),
    'siddhivinayak temple': (19.0176, 72.8308),
    'mahalaxmi railway station': (18.9827, 72.8235),
    'saifee hosptial': (18.9587, 72.8162),
    'hanging garden-mumbai': (18.9593, 72.8134),
    'tarapore aquarium': (18.9274, 72.8260),
    'wankhede stadium': (18.9388, 72.8255),
    'churney road railway station': (18.9575, 72.8040),
    'trident hotel': (18.9253, 72.8207),
    'kokilaben hospital': (19.1318, 72.8294),
    'borivali national park': (19.2295, 72.8543),
    'Nesco-goregaon': (19.1553, 72.8556)
}

adjacency_matrix_haversine = [[ 0 for _ in range(len(locations))] for _ in range(len(locations))]

for i, (place1, (latÄ±, lon1)) in enumerate(locations.items()):
  for j, (place2, (lat2, lon2)) in enumerate(locations.items()):
    if i != j:
      distance = haversine_distance (latÄ±, lon1, lat2, lon2)
      adjacency_matrix_haversine[i][j] = distance
hs=[]

for row in adjacency_matrix_haversine:
  hs.append(row)
```

```python
import heapq

start_point = 'DJ Sanghvi college'
end_point = 'trident hotel'

visited = set()
distances = {location: float('inf') for location in locations}
previous_locations = {location: None for location in locations}
distances[start_point] = 0
priority_queue = [(0, start_point)]

while priority_queue:
    current_distance, current_point = heapq.heappop(priority_queue)
    if current_point == end_point:
        break
    if current_point in visited:
        continue
    visited.add(current_point)
    for neighbor, distance in enumerate(adjacency_matrix_haversine[list(locations.keys()).index(current_point)]):
        if distance > 0:
            distance_to_neighbor = current_distance + distance

            if distance_to_neighbor < distances[list(locations.keys())[neighbor]]:
                distances[list(locations.keys())[neighbor]] = distance_to_neighbor
                previous_locations[list(locations.keys())[neighbor]] = current_point
                heapq.heappush(priority_queue, (distance_to_neighbor, list(locations.keys())[neighbor]))

shortest_path = []
current_point = end_point
while current_point:
    shortest_path.insert(0, current_point)
    current_point = previous_locations[current_point]

for i in range(len(shortest_path) - 1):
    print(f"{shortest_path[i]} <--- {shortest_path[i + 1]}")
```
