```python
def h1(table, goal_table):
    score = 0
    for i, row in enumerate(table):
        if i < len(goal_table):
            for j, block in enumerate(row):
                if j < len(goal_table[i]) and block == goal_table[i][j]:
                    score += 1
                else:
                    score -= 1
    score -= max(0, len(table) - len(goal_table))
    return score
```

```python
import random
def h1(table, goal_table):
    score = 0
    for i, row in enumerate(table):
        if i < len(goal_table):
            for j, block in enumerate(row):
                if j < len(goal_table[i]) and block == goal_table[i][j]:
                    score += 1
                else:
                    score -= 1
    score -= max(0, len(table) - len(goal_table))
    return score
def generate_neighbor(table, goal_table):
    neighbor = [list(row) for row in table]
    i = random.randint(0, len(neighbor) - 1)
    if not neighbor[i]:
        return neighbor

    j = random.randint(0, len(neighbor[i]) - 1)
    ni = random.randint(0, len(goal_table) - 1)
    while len(neighbor[ni]) >= len(goal_table[ni]):
        ni = random.randint(0, len(goal_table) - 1)

    block = neighbor[i].pop(j)
    neighbor[ni].append(block)

    return neighbor

def hill_climbing(initial_table, goal_table):
    current_table = initial_table
    current_score = h1(current_table, goal_table)
    steps = 0

    while True:
        neighbor = generate_neighbor(current_table, goal_table)
        neighbor_score = h1(neighbor, goal_table)
        steps += 1

        print(f"Step {steps}: Heuristic Score = {neighbor_score}")

        if neighbor_score >= current_score:
            current_table = neighbor
            current_score = neighbor_score
        else:
            break

    return current_score

if __name__ == "__main__":
    print("Enter the initial state:")
    initial_table = []
    for stack_num in range(1, 3):
        stack = input_blocks(stack_num)
        initial_table.append(stack)

    print("\nEnter the goal state:")
    goal_table = []
    for stack_num in range(1, 3):
        stack = input_blocks(stack_num)
        goal_table.append(stack)

    final_score = hill_climbing(initial_table, goal_table)

    print(f"\nFinal Heuristic Score: {final_score}")
```

```python
class BlocksWorld:
    def __init__(self, current_state):
        self.current_state = current_state

    def heuristic_1(self, goal_state):
        correct_blocks = 0
        for pile_index, pile in enumerate(goal_state):
            for block_index, block in enumerate(pile):
                if block in self.current_state[pile_index]:
                    correct_blocks += 1
                else:
                    correct_blocks -= 1
        return correct_blocks

    def heuristic_2(self, current_state):
        goal_state = [['D', 'C', 'B', 'E', 'A'], ['F']]
        n = sum(len(pile) for pile in goal_state)
        correct_blocks = 0
        for pile_index, pile in enumerate(goal_state):
            for block_index, block in enumerate(pile):
                if block in current_state[pile_index]:
                    correct_blocks += n
                else:
                    correct_blocks -= n
        return correct_blocks

def hill_climbing(problem, heuristic):
    current_state = problem.current_state
    goal_state = [['D', 'C', 'B', 'E', 'A'], ['F']]

    while True:
        neighbors = []
        for i in range(len(current_state)):
            for j in range(len(current_state[i])):
                for k in range(len(current_state)):
                    if k != i or len(current_state[k]) == 0:
                        new_state = [list(pile) for pile in current_state]
                        new_state[k].append(new_state[i].pop(j))
                        neighbors.append(new_state)

        best_neighbor = min(neighbors, key=lambda state: heuristic(state))

        if heuristic(best_neighbor) >= heuristic(current_state):
            return heuristic(best_neighbor)
        else:
            current_state = best_neighbor

def best_first_search(problem, heuristic):
    current_state = problem.current_state
    goal_state = [['D', 'C', 'B', 'E', 'A'], ['F']]

    priority_queue = [(heuristic(current_state), current_state)]

    while priority_queue:
        _, state = priority_queue.pop(0)

        if state == goal_state:
            return heuristic(state)

        neighbors = []
        for i in range(len(state)):
            for j in range(len(state[i])):
                for k in range(len(state)):
                    if k != i or len(state[k]) == 0:
                        new_state = [list(pile) for pile in state]
                        new_state[k].append(new_state[i].pop(j))
                        neighbors.append(new_state)

        priority_queue.extend([(heuristic(neighbor), neighbor) for neighbor in neighbors])

# (rest of the code remains the same)

# Define initial state
initial_state = [['D', 'C', 'B', 'A'], ['F', 'E']]

blocks_world = BlocksWorld(initial_state)

# Hill Climbing with Heuristic 1
final_value_hm1_hc = hill_climbing(blocks_world, blocks_world.heuristic_1)
print("Final Value (Hill Climbing, Heuristic 1):", final_value_hm1_hc)

# Hill Climbing with Heuristic 2
final_value_hm2_hc = hill_climbing(blocks_world, blocks_world.heuristic_2)
print("Final Value (Hill Climbing, Heuristic 2):", final_value_hm2_hc)

# Best First Search with Heuristic 1
final_value_hm1_bfs = best_first_search(blocks_world, blocks_world.heuristic_1)
print("Final Value (Best First Search, Heuristic 1):", final_value_hm1_bfs)

# Best First Search with Heuristic 2
final_value_hm2_bfs = best_first_search(blocks_world, blocks_world.heuristic_2)
print("Final Value (Best First Search, Heuristic 2):", final_value_hm2_bfs)
```
