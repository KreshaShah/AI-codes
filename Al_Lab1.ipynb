Water jug problem

```python
capacities = [3,5,8]
state = [8,0,0]
moves = []
num_jugs = len(capacities)
for i in range(num_jugs):
  for j in range(num_jugs):
    if i!=j:
      fill = min(state[i], capacities[j]- state[j])
      new_state = state.copy()
      new_state[i]-=fill
      new_state[j]+=fill
      moves.append(tuple(new_state))
print("Next possible moves are:")
for move in moves:
  print(move)
```

8-puzzle problem

```python
def gridPrint(blankX, blankY, tempX, tempY, board):
    for i in range(3):
        for j in range(3):
            if i == blankX and j == blankY:
                print(f"{board[tempX][tempY]}", end=" ")
            elif i == tempX and j == tempY:
                print(f"{board[blankX][blankY]}", end=" ")
            else:
                print(f"{board[i][j]}", end=" ")
        print()
board = []
posX = -1
posY = -1
solved = True
for i in range(3):
    row = []
    for j in range(3):
        num = int(input(f"Enter number for [{i+1}], [{j+1}]: "))
        row.append(num)
        if num != (i * 3) + (j + 1) % 9:
            solved = False
        if num == 0:
            posX, posY = i, j
    board.append(row)
if solved:
    print("Already solved")
else:
    for row in board:
        print(*row)
    print()
    moves = [(posX-1, posY, "up"), (posX, posY-1, "left"), (posX+1, posY, "down"), (posX, posY+1, "right")]
    for move in moves:
        x, y, direction = move
        if 0 <= x < 3 and 0 <= y < 3:
            print(f"If we move the blank space {direction}:")
            gridPrint(posX, posY, x, y, board)
```

TSP Problem

```python
cities = [1,2,3]
moves = []
from itertools import permutations
city_permutations = permutations(cities)
for perm in city_permutations:
  new_state = list(perm)
  if new_state != cities:
    moves.append(new_state)
print("Moves:",moves)
```
